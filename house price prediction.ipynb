**1. What is the average median income of the data set and check the
     distribution of data using appropriate plots. Please explain the
        distribution of the plot.**

```python
#importing pandas and matplotlib library
import pandas as pd
import matplotlib.pyplot as plt

#loading dataset
data=pd.read_excel('/content/housing+(1).xlsx')

#cal mean of median income in dataset
avg_med_income= data['median_income'].mean()
#print('average median income is:',avg_med_income.round(2))

#plotting histogram by using matplotli
plt.hist(data['median_income'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Median Income')
plt.ylabel('Frequency')
plt.title('Distribution of Median Income')

#adding a vertical line (axvline) at the position of the average median income. It is red colored dashed line,includes a label in the legend.
plt.axvline(avg_med_income, color='red',linestyle='dashed',label='Average Median Income(3.87)')
plt.legend()
plt.show()

#CONTINUOUS NUMERICAL VARIABLE
#OBSERVATION: distribution of average median income,median income on x-axis and frequency on y-axis & average median income lies between 2 and 4 and frequency is 3500
```


**2. Draw an appropriate plot to see the distribution of housing_median_age and explain your observations.**

```python
#visualizing the distribution using seaborn histograms
sns.histplot(data['housing_median_age'], bins=20, color='green', edgecolor='k'  )  # WE CAN USE KDE ALSO
#giving labels and titles
plt.xlabel('Housing Median Age')
plt.ylabel('Frequency')
plt.title('Distribution of Housing Median Age')
plt.axvline(data['housing_median_age'].mean(), color='red',linestyle='dashed',label='mean ')
plt.axvline(data['housing_median_age'].median(), color='green',label='median ')
plt.legend(fontsize='small',edgecolor= 'k')
plt.show()
#THIS DATA IS CONTINUOUS NUMERICAL DATA OR VAR

#OBSERVATION: distribution of housing medain age by median age on x-axis and frequency on y-axis & frequency is high between 30-40 age
```


**4. Create a data set by deleting the corresponding examples from the data set for which total_bedrooms are not available.**


**5. Create a data set by filling the missing data with the mean value of the total_bedrooms in the original data set.**

```python
#>>> here 1st we have to find missing data of entire dataset  by fillna() or isna() and fill it by mean of total bedrooms
missing_values = data.isna().sum()

#> calculating the mean of the 'total_bedrooms' column using data['total_bedrooms'].mean() and store it in the 'mean_tot_bedrooms' var.
mean_tot_bedrooms=data['total_bedrooms'].mean()

#> filling in missing values in the 'total_bedrooms' column with the calculated mean using the fillna method. **inplace=true make changes permanent
data['total_bedrooms'].fillna(mean_tot_bedrooms,inplace=True)
#  (or) count_mean_values=(data['total_bedrooms']==mean_tot_bedrooms).sum()
data

#DESCRIPTION: here i created a dataset by filling all missing values of total bedrooms by mean values of total bedrooms
```


6**. Write a programming construct (create a user defined function) to calculate the median value of the data set wherever required.**

```python
#> creating user def function
def cal_median(input_data):
    dataset = pd.read_excel(input_data)   #loading dataset
    median_values = dataset.median()      #calculating median
    return median_values                  # returnung medain value

cal_median( '/content/housing+(1).xlsx' )    #calling function

#DESCRIPTION: created a user defined function to calculate median values in a dataset
```


**10**. **Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small. If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered large.**


```python
#Use the .loc method to assign values to the 'total_bedroom_size' column based on conditions for the 'total_bedrooms' column:
data.loc[data['total_bedrooms'] <= 10, 'total_bedroom_size'] = 'small'
data.loc[(data['total_bedrooms'] > 10) & (data['total_bedrooms'] < 1000), 'total_bedroom_size'] = 'medium'
data.loc[data['total_bedrooms'] >= 1000, 'total_bedroom_size'] = 'large'
data

#DESCRIPTION: here i created a new column of total bedroom size of small,med,large by using loc method for selecting data from dataset
```
